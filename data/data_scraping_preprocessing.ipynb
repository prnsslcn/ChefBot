```python
# 메뉴명 추출
import os
import pandas as pd
    
# 메뉴 데이터가 있는 폴더 경로 설정
BASE_DIR = os.getcwd()  # 현재 작업 디렉토리
data_path = os.path.normpath(os.path.join(BASE_DIR, "./menu"))  # 상대경로 정리
menu_folder = os.path.join(data_path)

# CSV 파일 목록 가져오기
csv_files = [f for f in os.listdir(menu_folder) if f.endswith('.csv')]

# 여러 개의 CSV 파일을 하나의 DataFrame으로 합치기
menu_data_list = []
for file in csv_files:
    file_path = os.path.join(menu_folder, file)
    df = pd.read_csv(file_path, encoding='cp949') # 한글 인코딩
    df['source_file'] = file 
    menu_data_list.append(df)

# 모든 데이터를 하나의 DataFrame으로 합치기
combined_menu_data = pd.concat(menu_data_list, ignore_index=True)

# 데이터의 MENU_NM 컬람 값으로 리스트 만들기
menu_list = combined_menu_data['MENU_NM'].tolist()

# 최신 메뉴.txt 데이터 리스트에 삽입
with open(f'{data_path}/popular_menu_these_days.txt', 'r', encoding='utf-8') as f:
    temp_list = ([line.strip() for line in f.readlines()])  
    menu_list.extend(temp_list)

# 중복 제거
unique_menu_list = list(set(menu_list))

# 중복 제거한 리스트 개수 출력
print(f'중복 제거한 메뉴 개수: {len(unique_menu_list)}')

# 중복 제거한 리스트 일부 출력
unique_menu_list[:10]

```

```python
import requests, json, os, time, random
from bs4 import BeautifulSoup

# 현재 작업 디렉토리 확인 및 상대 경로 정리
BASE_DIR = os.getcwd()
data_path = os.path.normpath(os.path.join(BASE_DIR, "./recipes"))
os.makedirs(data_path, exist_ok=True)

print(f"데이터 저장 경로: {data_path}")

# 메뉴명을 입력하면 음식 정보를 출력하는 함수 (만개의 레시피 웹 스크래핑)
# 다수 요청시 계속 원격 중단되어 재시도 로직 추가
def food_info(name):
    recipe_list = []
    url = f"https://www.10000recipe.com/recipe/list.html?q={name}"

    # HTTP 요청 실패 시 재시도 (최대 3회)
    for attempt in range(3):
        try:
            response = requests.get(url, timeout=10)  
            if response.status_code == 200:
                break
            else:
                print(f"HTTP 응답 에러 ({response.status_code}), 재시도 {attempt+1}/3")
        except requests.RequestException as e:
            print(f"요청 실패: {e}, 재시도 {attempt+1}/3")
        time.sleep(random.uniform(3, 6)) 

    if response.status_code != 200:
        print(f"요청 실패: {url}")
        return None

    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    # 첫 번째 검색 결과에서 레시피 링크 가져오기
    food_list = soup.find_all(attrs={'class': 'common_sp_link'})
    if not food_list:
        print(f"레시피 검색 결과 없음: {name}")
        return None

    # 레시피는 최대 5개 저장
    for food in food_list[:5]:
        food_id = food['href'].split('/')[-1]
        new_url = f'https://www.10000recipe.com/recipe/{food_id}'

        # HTTP 요청 실패 시 재시도 (최대 3회)
        for attempt in range(3):
            try:
                new_response = requests.get(new_url, timeout=10)
                if new_response.status_code == 200:
                    break
                else:
                    print(f"레시피 페이지 요청 실패 ({new_response.status_code}), 재시도 {attempt+1}/3")
            except requests.RequestException as e:
                print(f"요청 실패: {e}, 재시도 {attempt+1}/3")
            time.sleep(random.uniform(3, 6))

        if new_response.status_code != 200:
            print(f"레시피 페이지 요청 실패: {new_url}")
            continue

        html = new_response.text
        soup = BeautifulSoup(html, 'html.parser')

        # JSON 데이터 추출
        food_info = soup.find(attrs={'type': 'application/ld+json'})

        # JSON 데이터가 없으면 건너뛰기
        if food_info is None:
            print(f"JSON 데이터 없음: {new_url}")
            continue

        try:
            result = json.loads(food_info.text)  # 🔹 JSON 파싱 시도
        except json.JSONDecodeError:
            print(f"JSON 파싱 실패 (형식 오류): {new_url}")
            continue

        # 'recipeIngredient'와 'recipeInstructions'가 있는지 확인
        if 'recipeIngredient' in result and 'recipeInstructions' in result:
            name = result['name']
            ingredient = ', '.join(result['recipeIngredient'])
            recipe = [step['text'] for step in result['recipeInstructions']]
            recipe = [f'{i+1}. {step}' for i, step in enumerate(recipe)]

            recipe_list.append({
                'name': name,
                'ingredients': ingredient,
                'recipe': recipe
            })

    # 3개 모두 정보가 없으면 "정보 없음" 처리
    if not recipe_list:
        recipe_list.append({
            'name': name,
            'ingredients': "정보 없음",
            'recipe': ["정보 없음"]
        })

    return recipe_list

# json 데이터를 파일로 저장하는 함수
def save_json(data, file_name):
    try:
        with open(file_name, 'w', encoding='utf-8') as file:
            json.dump(data, file, ensure_ascii=False, indent='\t')
            file.flush()
    except Exception as e:
        print(f"파일 저장 오류: {e}")

# 메뉴명 리스트로 레시피 생성 (딜레이 추가)
for i, menu in enumerate(national_dish_list):
    print(f"\n({i+1}/{len(national_dish_list)}) {menu} 레시피 검색 중...")

    res = food_info(menu)
    if res:
        safe_file_name = menu.replace(" ", "_").replace("/", "_").replace("\\", "_")
        file_path = os.path.join(data_path, f"{safe_file_name}_recipe.json")
        save_json(res, file_path)
        print(f"저장 완료: {file_path}")
    else:
        print(f'{menu}: 레시피를 찾을 수 없습니다.')

    # 요청 속도를 줄이기 위해 랜덤 대기 (3~6초)
    time.sleep(random.uniform(3, 6))
```
