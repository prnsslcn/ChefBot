```python
# 메뉴명 추출
import os
import pandas as pd

# # 현재 작업 디렉토리 확인 및 위치 조정 (현재 data 디렉토리에 있다는 가정 하에 진행)
if os.getcwd().endswith('data\menu'):
    print('이미 데이터 메뉴 폴더에 있습니다.')
else:
    os.chdir('./menu')
    print('데이터 메뉴 폴더로 이동했습니다.')
    
# 메뉴 데이터가 있는 폴더 경로 설정
current_dir = os.getcwd()
menu_folder = os.path.join(current_dir)

# CSV 파일 목록 가져오기
csv_files = [f for f in os.listdir(menu_folder) if f.endswith('.csv')]

# 여러 개의 CSV 파일을 하나의 DataFrame으로 합치기
menu_data_list = []
for file in csv_files:
    file_path = os.path.join(menu_folder, file)
    df = pd.read_csv(file_path, encoding='cp949') # 한글 인코딩
    df['source_file'] = file 
    menu_data_list.append(df)

# 모든 데이터를 하나의 DataFrame으로 합치기
combined_menu_data = pd.concat(menu_data_list, ignore_index=True)

# 데이터의 MENU_NM 컬람 값으로 리스트 만들기
menu_list = combined_menu_data['MENU_NM'].tolist()

# 중복 제거
unique_menu_list = list(set(menu_list))

# 유행하는 음식 저장 (따로 데이터 삽입)
unique_menu_list.append('두바이 초콜릿')
unique_menu_list.append('허니버터칩')
unique_menu_list.append('마라탕')
unique_menu_list.append('마라샹궈')
unique_menu_list.append('수건케이크')
unique_menu_list.append('요아정')

# 중복 제거한 리스트 개수 출력
print(f'중복 제거한 메뉴 개수: {len(unique_menu_list)}')

# 중복 제거한 리스트 일부 출력
unique_menu_list

```

```python
# 메뉴명으로 레시피 생성
import requests, json
from bs4 import BeautifulSoup

# 현재 작업 디렉토리 확인 및 위치 조정 (현재 data 레시피 폴더에 있다는 가정 하에 진행)
if os.getcwd().endswith('data\recipes'):
    print('이미 데이터 레시피 폴더에 있습니다.')
else:
    os.chdir('..')
    os.chdir('./recipes')
    print('데이터 레시피 폴더로 이동했습니다.')

# 메뉴명을 입력하면 음식 정보를 출력하는 함수 (만개의 레시피 웹 스크래핑)
def food_info(name):
    recipe_list = []
    url = f"https://www.10000recipe.com/recipe/list.html?q={name}"
    response = requests.get(url)
    if response.status_code == 200:
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')
    else: 
        print(f"HTTP 응답 에러 ({response.status_code}): {url}")
        return None

    # 첫 번째 검색 결과에서 레시피 링크 가져오기
    food_list = soup.find_all(attrs={'class': 'common_sp_link'})
    if not food_list:
        print(f"레시피 검색 결과 없음: {name}")
        return None
    
    # 레시피는 한 메뉴당 3개로 지정 (추후 적으면 더 늘릴 예정)
    # json 데이터 생성 중 예외 처리 안 해주면 반복문이 중단됨 
    for food in food_list[:3]:
        food_id = food['href'].split('/')[-1]
        new_url = f'https://www.10000recipe.com/recipe/{food_id}'
        new_response = requests.get(new_url)
        if new_response.status_code != 200:
            print(f"레시피 페이지 요청 실패 ({new_response.status_code}): {new_url}")
            return None

        html = new_response.text
        soup = BeautifulSoup(html, 'html.parser')

        # JSON 데이터 추출
        food_info = soup.find(attrs={'type': 'application/ld+json'})
        if food_info is None:
            print(f"JSON 데이터 없음: {new_url}")
            return None

        result = json.loads(food_info.text)

        # 'recipeIngredient' 키 확인 후 추출
        if 'recipeIngredient' in result:
            ingredient = ','.join(result['recipeIngredient'])
        else:
            print(f"'recipeIngredient' 키 없음: {new_url}")
            ingredient = "정보 없음"

        # 'recipeInstructions' 키 확인 후 추출
        if 'recipeInstructions' in result:
            recipe = [step['text'] for step in result['recipeInstructions']]
            recipe = [f'{i+1}. {step}' for i, step in enumerate(recipe)]
        else:
            print(f"'recipeInstructions' 키 없음: {new_url}")
            recipe = ["정보 없음"]

        res = {
            'name': name,
            'ingredients': ingredient,
            'recipe': recipe
        }

        recipe_list.append(res)

    return recipe_list

# json 데이터을 파일로 저장하는 함수
def save_json(data, file_name):
    # 파일 덤프
    with open(file_name, 'w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent='\t')

# 메뉴명 리스트로 레시피 생성
for i in range(len(unique_menu_list)):
    res = food_info(unique_menu_list[i])
    print(res)
    if res:
        save_json(res, f"{res[0]['name']}_recipe.json")  # 딕셔너리에서 'name' 키 사용
    else:
        print(f'{unique_menu_list[i]}: 레시피를 찾을 수 없습니다.')

```
